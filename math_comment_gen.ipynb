```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from IPython.display import HTML

# Define widget components with adjusted layout
name_widget = widgets.Text(
    placeholder="Enter student name",
    description="Name:",
    style={"description_width": "150px"},
    layout=widgets.Layout(width="750px"),
)
pronoun_widget = widgets.Dropdown(
    options=["he", "she", "they"],
    description="Pronoun:",
    style={"description_width": "150px"},
    layout=widgets.Layout(width="750px"),
)
computation_widget = widgets.Text(
    placeholder="Enter computation terms",
    description="Computation:",
    style={"description_width": "150px"},
    layout=widgets.Layout(width="750px"),
)
optional_toggle = widgets.Checkbox(
    value=False,
    description="Optional Section",
    style={"description_width": "150px"},
)

# Dropdowns for categories and sentences with wider layout
# concepts_category_widget = widgets.Dropdown(
#     options=["Strengths", "Progressing", "Challenges", "New Student"],
#     description="Category:",
#     style={"description_width": "150px"},
#     layout=widgets.Layout(width="750px"),
# )

concepts_category_widget = widgets.Dropdown(
    options=["Select a category", "Strengths", "Progressing", "Challenges", "New Student"],
    description="Category:",
    style={"description_width": "150px"},
    layout=widgets.Layout(width="750px"),
)

concepts_sentence_widget = widgets.Dropdown(
    options=[], description="Sentence:", style={"description_width": "150px"}, layout=widgets.Layout(width="750px")
)

computation_category_widget = widgets.Dropdown(
    options=["Strengths", "Progressing", "Challenges"],
    description="Category:",
    style={"description_width": "150px"},
    layout=widgets.Layout(width="750px"),
)
computation_sentence_widget = widgets.Dropdown(
    options=[], description="Sentence:", style={"description_width": "150px"}, layout=widgets.Layout(width="750px")
)

problem_solving_category_widget = widgets.Dropdown(
    options=["Strengths", "Progressing", "Challenges", "New Student"],
    description="Category:",
    style={"description_width": "150px"},
    layout=widgets.Layout(width="750px"),
)
problem_solving_sentence_widget = widgets.Dropdown(
    options=[], description="Sentence:", style={"description_width": "150px"}, layout=widgets.Layout(width="750px")
)

optional_category_widget = widgets.Dropdown(
    options=["Strengths", "Challenges"],
    description="Optional Category:",
    style={"description_width": "150px"},
    layout=widgets.Layout(width="750px"),
)
optional_sentence_widget = widgets.Dropdown(
    options=[], description="Optional Sentence:", style={"description_width": "150px"}, layout=widgets.Layout(width="750px")
)

closing_sentence_widget = widgets.Dropdown(
    options=closing_statements,
    description="Closing Sentence:",
    style={"description_width": "150px"},
    layout=widgets.Layout(width="750px"),
)

# Introductory statement
introductory_widget = widgets.Dropdown(
    options=introductory_statements,
    description="Introductory Statement:",
    style={"description_width": "150px"},
    layout=widgets.Layout(width="750px"),
)

# Dynamic sentence update
def update_sentence_options(change):
    pronoun = pronoun_widget.value
    concepts_sentence_widget.options = sections_dict["Concepts"][concepts_category_widget.value][pronoun]
    computation_sentence_widget.options = sections_dict["Computation"][computation_category_widget.value][pronoun]
    problem_solving_sentence_widget.options = sections_dict["Problem Solving"][problem_solving_category_widget.value][pronoun]
    optional_sentence_widget.options = (
        sections_dict["Optional"][optional_category_widget.value][pronoun]
        if optional_toggle.value
        else []
    )

# Bind updates
pronoun_widget.observe(update_sentence_options, names="value")
concepts_category_widget.observe(update_sentence_options, names="value")
computation_category_widget.observe(update_sentence_options, names="value")
problem_solving_category_widget.observe(update_sentence_options, names="value")
optional_category_widget.observe(update_sentence_options, names="value")
optional_toggle.observe(update_sentence_options, names="value")

# Output area
output_area = widgets.Output()

def generate_paragraph(*args):
    with output_area:
        clear_output()
        name = name_widget.value
        pronoun = pronoun_widget.value
        computation_text = computation_widget.value
        
        intro_text = introductory_widget.value.replace("_______", name)
        concept_text = concepts_sentence_widget.value.replace("_______", name)
        computation_text_final = computation_sentence_widget.value.replace("_______", name).replace("-----", computation_text)
        problem_solving_text = problem_solving_sentence_widget.value.replace("_______", name)
        optional_text = (
            optional_sentence_widget.value.replace("_______", name) if optional_toggle.value else ""
        )
        closing_text = closing_sentence_widget.value.replace("_______", name)
        
        paragraph = (
            f"{intro_text} {concept_text} {computation_text_final} {problem_solving_text} "
            f"{optional_text} {closing_text}".strip()
        )

        # handle the case where a name ends in an s
        if name.endswith("s"):
            paragraph = paragraph.replace(f"{name}'s", f"{name}'")
        
        # Wrap the paragraph in an HTML div for proper styling
        styled_paragraph = f"""
        <div style="word-wrap: break-word; white-space: pre-wrap; width: 750px;">
            {paragraph}
        </div>
        """
        display(HTML(styled_paragraph))

# Update button
update_button = widgets.Button(description="Generate")
update_button.on_click(generate_paragraph)

# Section titles
info_title = widgets.HTML(value="<b>Basic Information</b>")
intro_title = widgets.HTML(value="<b>Intro Sentence</b>")
concepts_title = widgets.HTML(value="<b>Concepts Section</b>")
computation_title = widgets.HTML(value="<b>Computation Section</b>")
problem_solving_title = widgets.HTML(value="<b>Problem Solving Section</b>")
optional_title = widgets.HTML(value="<b>Optional Section</b>")
closing_title = widgets.HTML(value="<b>Closing Sentence</b>")
click_title = widgets.HTML(value="<b>Click Below to Generate</b>")


# Layout widgets
widgets_box = widgets.VBox([
    info_title,
    name_widget,
    pronoun_widget,
    intro_title,
    introductory_widget,
    concepts_title,
    concepts_category_widget,
    concepts_sentence_widget,
    computation_title,
    computation_widget,
    computation_category_widget,
    computation_sentence_widget,
    problem_solving_title,
    problem_solving_category_widget,
    problem_solving_sentence_widget,
    optional_title,
    optional_toggle,
    optional_category_widget,
    optional_sentence_widget,
    closing_title,
    closing_sentence_widget,
    click_title,
    update_button,
    output_area,
])

# Display
display(widgets_box)
```
